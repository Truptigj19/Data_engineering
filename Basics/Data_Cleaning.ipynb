```python
def clean_transactions(data):
    import pandas as pd
    
    df = pd.DataFrame(data, columns=["amount"])
    
    # 1. Remove null-like values
    df = df[~df["amount"].isin([None, "", "NULL"])]
    
    # 2. Convert to numeric
    df["amount"] = pd.to_numeric(df["amount"], errors="coerce")
    
    # 3. Remove invalid (NaN)
    df = df.dropna()
    
    # 4. Apply business rule (>0)
    df = df[df["amount"] > 0]
    
    # 5. Remove duplicates
    df = df.drop_duplicates()
    
    # 6. Sort data
    df = df.sort_values(by="amount")
    
    return df["amount"].tolist()
```

```python
data = [100, -50, None, 200, "300", 100, -10, "invalid", 400,
        "500.5", 0, "", "700", -300, 200, "NULL"]

print(clean_transactions(data))
```
