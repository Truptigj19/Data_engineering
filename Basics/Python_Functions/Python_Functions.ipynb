Q3. Write transformation Function for: 
Convert amount to int/float
Keep only: {"customer": ..., "amount": ...} 
	(Schema Standardization)

Expected Approach
	Copy data (don’t mutate original)
	Convert safely

```python
def transform_all(valid_record):
    final_data =[]
    for t in valid_record:
        new_record = {
            "customer": t["customer"],
             "amount": float(t["amount"])
        }
        final_data.append(new_record)

    return final_data


result = transform_all(valid_record)
for item in result:
    print(item)
```
