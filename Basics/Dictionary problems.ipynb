# Dictionary problems 


```python
transactions = [
    {"id": 1, "customer": "A", "amount": 100, "status": "SUCCESS"},
    {"id": 2, "customer": "B", "amount": "200", "status": "FAILED"},
    {"id": 3, "customer": "A", "amount": None, "status": "SUCCESS"},
    {"id": 4, "customer": "C", "amount": 300, "status": "SUCCESS"},
    {"id": 5, "customer": "B", "amount": "invalid", "status": "SUCCESS"},
    {"id": 6, "customer": "A", "amount": 100, "status": "SUCCESS"},
    {"id": 7, "customer": "D", "amount": -50, "status": "SUCCESS"},
]
```

Questions: 

Q1. Filter Successful Transactions
👉 Keep only records where:	status == "SUCCESS"
Pattern: Filtering

```python
success_txn  = [t for t in transactions if t["status"] == "SUCCESS" ]
success_txn
```

Q2. Clean Amount Field
👉 Convert: "200" → 200
👉 Remove: None and "invalid"
Pattern: Transformation + validation

```python
clean_txn = []
invalid_txn = []
for t in success_txn:
    amt = t["amount"]
    try:
        amt = int(amt)
        t["amount"] = amt
        clean_txn.append(t)
    except:
        invalid_txn.append(t)

print("Cleaned Transcation")
clean_txn
```

```python
print("Invalid Transcation")
invalid_txn
    
```

Q3. Apply Business Rule
👉 Keep only transactions where: amount > 0
Pattern: Business filtering


```python
final_txn = [t for t in clean_txn if t["amount"] > 0]
final_txn
```

Q4. Extract Only Required Fields
👉 Create new list with only: {"customer": ..., "amount": ...}
Pattern: Data projection


```python
projected_txn = []
for t in final_txn:
    new_record = {
        "customer" : t['customer'],
        'amount' : t['amount']
    }
    projected_txn.append(new_record)
projected_txn
```

Q5. Remove Duplicate Transactions
👉 Duplicate logic:	Same customer + same amount
Pattern: Deduplication

```python
seen = set()
unique_txn = []

for t in projected_txn:
    key = (t['customer'],t['amount'])
    if key not in seen:
        seen.add(key)
        unique_txn.append(t)

unique_txn


        
```

Q6. Total Amount per Customer
👉 Output:
{
    "A": 200,
    "B": 200,
    "C": 300
}
Pattern: Aggregation (GROUP BY)

```python
total_per_customer = {}
for t in projected_txn:
    cust = t["customer"]
    amt = t["amount"]
    if cust in total_per_customer:
        total_per_customer[cust] += amt
    else:
        total_per_customer[cust] = amt

total_per_customer
        
        
```

Q7. Count Transactions per Customer
👉 Output:
{
    "A": 3,
    "B": 1,
    "C": 1
}
Pattern: Counting / aggregation

```python
count_per_customer = {}

for t in projected_txn:
    cust = t["customer"]
    
    if cust in count_per_customer:
        count_per_customer[cust] += 1
    else:
        count_per_customer[cust] = 1

count_per_customer 
```

Q8. Find Top Customer 
👉 Customer with highest total amount
Pattern: Aggregation + max


```python
max_value = 0
top_customer = None

for cust, amt in total_per_customer.items():
    if amt > max_value:
        max_value = amt
        top_customer = cust

top_customer
```

    return result

👉 Should:
	Filter SUCCESS
	Clean amount
	Apply business rules
	Remove duplicates
	Return total amount per customer

Expected Final Output: 
{
    "A": 200,
    "C": 300
}

```python
def process_transactions(data):
    
    # Step 1: Filter SUCCESS
    success = [t for t in data if t["status"] == "SUCCESS"]
    
    # Step 2: Clean amount + apply business rule (>0)
    clean = []
    for t in success:
        try:
            amt = int(t["amount"])
            if amt > 0:
                clean.append({
                    "customer": t["customer"],
                    "amount": amt
                })
        except:
            continue
    
    # Step 3: Remove duplicates (customer + amount)
    seen = set()
    unique = []
    
    for t in clean:
        key = (t["customer"], t["amount"])
        if key not in seen:
            seen.add(key)
            unique.append(t)
    
    # Step 4: Aggregate (total per customer)
    result = {}
    
    for t in unique:
        cust = t["customer"]
        result[cust] = result.get(cust, 0) + t["amount"]
    
    return result


print(process_transactions(transactions))
```
