# Dict sorting and aggregation

```python
orders = [
    {"order_id": 1, "customer": "A", "amount": 100, "city": "Pune"},
    {"order_id": 2, "customer": "B", "amount": 200, "city": "Mumbai"},
    {"order_id": 3, "customer": "A", "amount": 150, "city": "Pune"},
    {"order_id": 4, "customer": "C", "amount": 300, "city": "Delhi"},
    {"order_id": 5, "customer": "B", "amount": 50, "city": "Mumbai"},
    {"order_id": 6, "customer": "A", "amount": 200, "city": "Pune"},
]
```

Q1. Sort Orders by Amount (Descending)
👉 Return full list sorted by amount (highest first)
Pattern: Sorting with key


```python
sorted_orders = sorted(orders, key = lambda x: x['amount'] , reverse=True)
sorted_orders
```

Q2. Top 2 Highest Orders 
👉 Return top 2 orders based on amount
Pattern: Sort + slice


```python
top_2 = sorted_orders[:2]
top_2
```

Q3. Second Highest Order 
👉 Return only the second highest order
👉 Edge case: if same values exist?
Pattern: Sorting + indexing

```python
sorted_orders[1:2]
```

Q4. Group Orders by Customer (Structured Output) 
👉 Output:
{
    "A": [100, 150, 200],
    "B": [200, 50],
    "C": [300]
}
👉 Only store amounts (not full dict)
Pattern: Grouping + transformation

```python
cust_amount = {}
for t in orders:
    customer  = t["customer"]
    if customer not in cust_amount:
        cust_amount[customer] = []
    cust_amount[customer].append(t["amount"])

cust_amount
        
```

Q5. Customer-wise Sorted Amounts ⭐⭐
👉 Output:
{
    "A": [200, 150, 100],
    "B": [200, 50],
    "C": [300]
}
👉 Sort each customer’s data (descending)
Pattern: Group + sort

```python
sorted_cust_amt = {}
for t in cust_amount:
    sorted_cust_amt[t] = sorted(cust_amount[t], reverse=True)    

sorted_cust_amt
        
```

Q6. Filter High-Value Orders
👉 Return orders where: amount > 150
Pattern: Filtering with condition


```python
high_value_cust = []
for t in orders:
    if t['amount'] > 150:
        high_value_cust.append(t)
high_value_cust
```

Q7. Multi-Condition Filter 
👉 Return orders where: city == "Pune" AND amount > 120
Pattern: Multiple conditions

```python
high_value_cust = []
for t in orders:
    if t['amount'] > 120 and t['city'] == 'Pune':
        high_value_cust.append(t)
high_value_cust
```

Q8. Create Simple Report Format
👉 Output:
[
    {"customer": "A", "total_orders": 3, "total_amount": 450},
    {"customer": "B", "total_orders": 2, "total_amount": 250},
    {"customer": "C", "total_orders": 1, "total_amount": 300}
]

```python
report = {}

for t in orders:
    customer = t["customer"]
    amount = t["amount"]

    if customer not in report:
        report[customer] = {
            "total_orders": 0,
            "total_amount": 0
        }

    report[customer]["total_orders"] += 1
    report[customer]["total_amount"] += amount


final_report = []

for customer in report:
    final_report.append({
        "customer": customer,
        "total_orders": report[customer]["total_orders"],
        "total_amount": report[customer]["total_amount"]
    })

print(final_report)
```
